```python
import pandas as pd
import numpy as np
```

```python
file_source='https://raw.githubusercontent.com/RaterBai/DS-Project/master/prepared%20datasets/'
```

```python
demo=pd.read_csv(file_source+'demographics.csv',error_bad_lines=False)
```

```python
demo
```

```python
demo=demo.rename(columns={'year_of_birth':'year','month_of_birth':'month','day_of_birth':'day'})
```

```python
demo['birth_date']=pd.to_datetime(demo[['year','month','day']])
```

```python
demo['first_visit_date']=pd.to_datetime(demo['first_visit_date'])
```

```python
demo['age']=(demo['first_visit_date']-demo['birth_date'])/np.timedelta64(1,'Y')
```

```python
demo
```

```python
demo=demo.drop(columns=['year','month','day','first_visit_date','birth_date'])
```

```python
demo['age'] = demo['age'].apply(np.int64)
```
